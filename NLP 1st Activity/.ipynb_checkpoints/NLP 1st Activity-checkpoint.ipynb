```python
import os
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
from wordcloud.wordcloud import WordCloud
import re
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import matplotlib.style

%matplotlib inline
```

```python
'''
Importing Raw data
'''
text=open('ref.txt',"r")
data=text.read()
```

```python
'''
Assign ENSW the stopwords
'''
ensw=stopwords.words('english')
```

```python
'''
Tokenize textArray
'''
textArray=word_tokenize(data.lower())
```

```python
'''
Return a list of filter words compared to the stopwords
'''
filterArray=[item for item in textArray if item not in ensw]
```

```python
eng_words=set(nltk.corpus.words.words())
```

```python
eng_filterArray=[word for word in filterArray if word in eng_words]
```

```python
'''
Creating Positive and Negative List
'''
PText=open('positive.txt','r')
positive=PText.read().split()

NText=open('negative.txt','r')
negative=NText.read().split()
```

```python
'''
Create 2 list of words
1. for_count_filter_positive: with duplication
2. filter positive: without duplication
'''
filter_positive=[word for word in eng_filterArray if word in positive]

```

```python
filter_positive
```

```python
'''
Create 2 list of words
1. for_count_filter_positive: with duplication
2. filter positive: without duplication
'''
filter_negative=[word for word in eng_filterArray if word in negative]
```

```python
'''
Count the words for preprocess raw data
'''
Counter(eng_filterArray)
```

```python
'''
Count the word of positive data
'''
Counter(filter_positive)
```

```python
'''
Count the word for negative data
'''
Counter(filter_negative)
```

```python
final_data=' '.join(set(eng_filterArray))
```

```python
positive_result=' '.join(set(filter_positive))
negative_result=' '.join(set(filter_negative))
```

```python
'''
Plotting Cloud for 
'''
plt.figure(figsize=(20,10))

wordcloud=WordCloud(background_color='white',mode='RGB',width=1800,height=1400).generate(final_data)
plt.title('Responses WordCloud')
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(20,10))
# postive_mask=np.array(Image.open('Plus_sign.png'))
wordcloud=WordCloud(background_color='white',mode='RGB',width=2000,height=1000).generate(positive_result)

plt.title('Positive Result')
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(20,10))
wordcloud=WordCloud(background_color='white',mode='RGB',width=2000,height=1000).generate(negative_result)
plt.title('Negative Result')
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
'''
Sentiment Analyzer
'''
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer=SentimentIntensityAnalyzer()
```

```python
filepath='ref.txt'
with open(filepath) as fp:
    line=fp.readline()
    cnt=1
    fa=open('PolarityList.txt','w')
    while line:
        data=line.strip()
        countList=list(analyzer.polarity_scores((data)).items())
        
        vs=':'.join([str(i) for i in countList])
        
        fa.write(vs + '\n')
        print(vs)
        line=fp.readline()
        cnt+=1
```
